```python
import random
import numpy as np
import seaborn as sns
```

<h1>ESTRATÉGIA 1 -> MANTÉM SEMPRE A PRIMEIRA ESCOLHA

```python
jogo_estrategia1 = []
prob1 = []
n_rodadas = 1000
for rodada in range(0, n_rodadas):
    #0 para cabra
    #1 para carro
    #depois embaralho tudo com shuffle para que a posição das cabras e do carro mude em cada rodada
    portas = [0,0,1]
    np.random.shuffle(portas)

    pos_portas = [0,1,2]
    escolha1 = random.choice(pos_portas)

    #1 GANHAR
    #0 PERDER
    if portas[escolha1] == 1:
        resultado = 1
        jogo_estrategia1.append(resultado)

    else:
        resultado = 0
        jogo_estrategia1.append(resultado)
    
    p = sum(jogo_estrategia1)/len(jogo_estrategia1)
    prob1.append(p)
```

```python
#sns.set_context("paper")
#sns.set_style("darkgrid")
sns.set_theme()
sns.despine(left=True)
#probabilidade de acertar X número de rodadas com estratégia 1
#perceba como a probabilidade tende a 1/3, isto é, a probabilidade de acertar a porta com carro na primeira escolha e não mudando-a
sns.lineplot(x = list(range(1,n_rodadas+1)), y = prob1).\
set(title='Estratégia 1: sempre manter primeira escolha', xlabel='Número de rodadas', ylabel='Probabilidade de sucesso')
```

<h1>Estratégia 2 --> sempre muda opção

```python
jogo_estrategia2 = []
prob2 = []
n_rodadas = 1000

for rodada in range(0,n_rodadas):
    portas = [0,0,1]
    np.random.shuffle(portas)

    pos_portas = [0,1,2]
    escolha1 = random.choice(pos_portas)

    if portas[escolha1] == 1:
        resultado = 0
        jogo_estrategia2.append(resultado)
    else:
        resultado = 1
        jogo_estrategia2.append(resultado)

    p = sum(jogo_estrategia2)/len(jogo_estrategia2)
    prob2.append(p)
```

```python
#probabilidade de acertar X número de rodadas com estratégia 2
#perceba como a probabilidade tende a 2/3, isto é, a probabilidade de acertar a porta com carro na segunda escolha
sns.lineplot(x = list(range(1,n_rodadas+1)), y = prob2).\
set(title='Estratégia 2: sempre mudar a primeira escolha', xlabel='Número de rodadas', ylabel='Probabilidade de sucesso')
```
